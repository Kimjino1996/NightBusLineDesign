```python
import pandas as pd
data=pd.read_csv("./bus_on.csv",encoding ="cp949")
data
```

```python
# 100단위로 검사하자.
test=[] # 위도
test2=[] # 경도
test3=[] # value 

for j in range(0,17):
    for i in range (0,30):
        x=round(35.9681-j*0.01,4)
        y=round(128.4651+i*0.01,4)
        test.append(x)
        test2.append(y)
        test3.append(0)
        
    
```

```python
from pandas import DataFrame
cluster=DataFrame(data={'lat':test,'lon':test2,'value':test3})
```

```python
cluster
```

```python
data['label']=-1
for i in range(0,len(data)):
    X=float(data.loc[i,'lat'])
    Y=float(data.loc[i,'lng'])
    
    for j in range(1,len(cluster)):
        lat=cluster.loc[j,'lat']
        lon=cluster.loc[j,'lon']
        if lat>X and (lat-0.01)<X and lon<Y and lon+0.01>Y:
            data.loc[i,'label']=j
    print("현재까지 :  ",i,"해당 i 의 value 값은 : ",data.loc[i,'label'])
```
