```python
import pandas as pd
data=pd.read_csv("./subway_off.csv",encoding="cp949")
data
```

```python
# 100단위로 검사하자.
test=[] # 위도
test2=[] # 경도
test3=[] # value 

for j in range(0,17):
    for i in range (0,30):
        x=round(35.9681-j*0.01,4)
        y=round(128.4651+i*0.01,4)
        test.append(x)
        test2.append(y)
        test3.append(0)
        
    
```

```python
from pandas import DataFrame
cluster=DataFrame(data={'lat':test,'lon':test2,'value':test3})
```

```python
cluster
```

```python
for i in range(0,510):
    lat=cluster.loc[i,'lat']
    lon=cluster.loc[i,'lon']
    for j in range(1,len(data)):
        X=float(data.loc[j,'lat'])
        Y=float(data.loc[j,'lng'])
        
        if lat>X and (lat-0.01)<X and lon<Y and lon+0.01>Y:
            print("해당 정보가 들어가고 있음.", i, lat, lon, X, Y, cluster.loc[i,'value'])
            cluster.loc[i,'value']=cluster.loc[i,'value']+data.loc[j,'total']
    print("현재까지 :  ",i,"해당 i 의 value 값은 : ",cluster.loc[i,'value'])
```

```python
for i in range(0,len(cluster)):
    print(cluster.loc[i,'value'])
```
