```python
import pandas as pd
import folium
import queue
import time
import math

cluster = pd.read_csv("total_taxi_cluster.csv")
```

```python
check_v = []

for i, row in cluster.iterrows():
    check_v.append(row['value'])
```

```python
# 위쪽에서 아래쪽으로의 이동 경로
# 이동 방향은 하, 좌, 우

def printMap(num):
    print("{}번째 이동".format(num))
    
    for i in range(17):
        for j in range(30):
            print("{0:3s}".format(visual_path[i][j]), end = '')
        print()
    
    print("\n\n")
    
    

def BFS(x,y):
    num = 1
    q.put((x,y))

    while q.qsize() != 0:
        findMax = []         # 가야할 방향의 노드들 위치를 저장하여 가장 큰 value 값을 가지는 노드를 리턴하기 위한 용도
        cur = q.get()        # cur : 현재 위치의 노드

        for d in direct:
            x = cur[0] + d[0]
            y = cur[1] + d[1]
            
            findMax.append((x, y))
        
        max_v = 0
        
        for path in findMax:
            
            if check_v[ground[path[0]][path[1]]] >= max_v:
                max_v = check_v[ground[path[0]][path[1]]]
                next_path = path    # 다음으로 이동할 노드의 좌표
                
        q.put(next_path)
        check_v[ground[next_path[0]][next_path[1]]] = -1
        visual_path[next_path[0]][next_path[1]] = '#'
        
        printMap(num)
        num+=1
        time.sleep(2)

        
q = queue.Queue()

direct = [(0, -1), (0, 1), (1, 0), (0, -2), (0, 2), (2, 0)]

ground = []
visual_path = [['0' for _ in range(30)] for _ in range(17)]

cnt = 0

for i in range(17):
    l = []
    
    for j in range(30):
        l.append(cnt)
        cnt += 1
    ground.append(l)
    
check_v[ground[2][9]] = -1
BFS(2, 9)
```
