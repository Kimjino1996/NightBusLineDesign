```python
import pandas as pd
data=pd.read_csv("C:/Users/황지현/Desktop/빅데이터 공모전/택시데이터정제/total_off_taxi_cluster.csv")
data
```

```python
import folium
map_osm = folium.Map(location=[35.8796687,128.4266147], zoom_start=13)
#card_data_place.dropna()

for i in range(0,len(data.index)):
    lat1=data.loc[i,'lat']
    lon1=data.loc[i,'lon']
    if data.loc[i,'value']>0:
        print(i,data.loc[i,'value'])
        folium.CircleMarker([lat1,lon1],radius=data.loc[i,'value']/2000,popup=i,color='red',fill=True).add_to(map_osm)   

map_osm.save('total_off_taxi_cluster.html')
```
