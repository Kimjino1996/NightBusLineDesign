```python
import pandas as pd
import folium
```

```python
data1=pd.read_csv("./bugu.csv",encoding='CP949')
data2=pd.read_csv("./dalseogu.csv",encoding='CP949')
data3=pd.read_csv("./dalseonggun.csv",encoding='CP949')
data4=pd.read_csv("./dongu.csv",encoding='CP949')
data5=pd.read_csv("./Jungu.csv",encoding='CP949')
data6=pd.read_csv("./namgu.csv",encoding='CP949')
data7=pd.read_csv("./sugu.csv",encoding='CP949')
data8=pd.read_csv("./susunggu.csv",encoding='CP949')
```

```python
total_data = [data1,data2,data3,data4,data5,data6,data7,data8]
```

```python
# 행정 구역, 2020년04월_거주자_인구수 col name

name = []
pop = []
lat = []
lng = []

for data in total_data:
#     print(len(data))
    for i, row in data.iterrows():
        name.append(row['행정구역'])
        pop.append(row['2020년04월_거주자 인구수'])
        lat.append(row['lat'])
        lng.append(row['lon'])
        
df = {'name' : name, 'cur_pop' : pop, 'lat':lat,'lng':lng}
final_data = pd.DataFrame(df)
final_data
```

```python
map_osm = folium.Map(location=[35.8796687,128.4266147], zoom_start=13)
```

```python
for i, row in final_data.iterrows():
    if int(row['cur_pop'].replace(',','')) > 30000:
        folium.CircleMarker([row['lat'],row['lng']], 
                            popup=row['name'],
                            radius=float(row['cur_pop'].replace(',',''))/1000, 
                            color='red').add_to(map_osm)
    else:
        folium.CircleMarker([row['lat'],row['lng']], 
                            popup=row['name'],
                            radius=float(row['cur_pop'].replace(',',''))/1000, 
                            color='blue').add_to(map_osm)
map_osm.save('map.html')
```
